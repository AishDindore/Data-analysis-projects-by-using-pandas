Student Exam Performance Analysis

Dataset: Student Performance (Kaggle)

Tasks:

Average scores by gender or parental level of education

Correlation between test prep and scores

Subject-wise top performers

pandas tools: corr(), groupby(), plot()

```python
import pandas as pd
import numpy as np
```

```python
stud_perform=pd.read_excel(r"C:\Users\Admin\Desktop\python\Minskole class of data analytics\pandas\practise datasets\Student_Exam_Performance_Analysis.xlsx")
stud_perform
```

Average scores by gender or parental level of education

```python
avg_score=stud_perform.groupby("Gender")["Score"].mean().reset_index()
avg_score
```

Subject-wise top performers

```python
Subject_wise_top = stud_perform.groupby("Subject")["Score"].max().reset_index().sort_values(by="Score",ascending=False)
Subject_wise_top
```

Add pass or fail column based on the score

```python
stud_perform["Pass/Fail"]=np.where(stud_perform["Score"]>=60,"Pass","Fail")
stud_perform
```

or method2 using lambda function

```python
stud_perform["Pass/Fail"]=stud_perform["Score"]>=60
```

```python


stud_perform["Pass/Fail"]=stud_perform["Score"].apply(lambda x: "Pass" if x >= 60 else "Fail")
stud_perform
```

Average Score per Subject

```python
Avg_score_subj=stud_perform.groupby("Subject")["Score"].mean().reset_index()
Avg_score_subj
```

```python

top_scorers =stud_perform.loc[stud_perform.groupby("Subject")["Score"].idxmax()][["Subject", "Name", "Score"]]
top_scorers
```

 Attendance vs. Score Correlation

```python
correlation=stud_perform["Attendance (%)"].corr(stud_perform["Score"])

print("Correlation between attendance and score:", correlation)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.scatterplot(data=stud_perform, x="Attendance (%)", y="Score", hue="Extra Classes Attended")
plt.title("Attendance vs Score")
plt.show()
```

 Impact of Extra Classes

```python
Extra_Classes=stud_perform.groupby("Extra Classes Attended")["Score"].mean().reset_index().sort_values(by="Score",ascending=False)
Extra_Classes

```
